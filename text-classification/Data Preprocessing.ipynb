```python
# clean data

import re
import nltk

def basic_clean(text):
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english')
    words = nltk.tokenize.word_tokenize(text)
    words = [w.lower() for w in words]
    temp = [word for word in words if re.search('^[a-z]+$', word)]
    temp2 = [wnl.lemmatize(word) for word in temp if word not in stopwords]
    return " ".join(temp2)
```

```python
df['text_aft'] = df['text'].map(basic_clean)
```

```python
df['title_aft'] = df['title'].map(basic_clean)
```

```python
# replace empty cell with NaN value
df = df.replace(r'^\s*$', np.NaN, regex=True)
```

```python
# check rows with null values
df[df.isnull().any(axis=1)]
```

```python
# drop original data columns to reduce volume to store
df = df.drop(['text', 'title'], axis=1)
```

```python
df = df[['title_aft', 'text_aft','subject', 'date', 'target']] 
```
